```python
apa['Price1'] = apa['Close'].shift(-1)
# close price of tomorrow
```

```python
apa['PriceDiff'] = apa['Price1']-apa['Close']
apa.head()
```

```python
#Daily return = price difference/ close price of today
apa['Return'] = apa['PriceDiff']/apa['Close']
apa.head()
```

```python
# Direction (PriceDiff>0 up, <0 down)
apa['Direction'] = [1 if apa.loc[ei,'PriceDiff']>0 else -1                          for ei in apa.index]
apa.head()
```

```python
# Moving average 
apa['MA3'] = (apa['Close']+apa['Close'].shift(1)+apa['Close'].shift(2))/3
```

# Strategy 
### When MA10 > MA50 buy and hold one share of stock

```python
apa['MA10'] = apa['Close'].rolling(10).mean() # fast signal
apa['MA50'] = apa['Close'].rolling(10).mean() # slow signal
# rolling window to
```

```python
apa['Close'].plot(legend=True,figsize=(20,8))
apa['MA50'].plot(legend=True,figsize=(20,8))
apa['MA10'].plot(legend=True,figsize=(20,8))
```

```python
apa['Close1'] = apa['Close'].shift(-1)
apa['Profit'] = [apa.loc[ei,'Close1']-apa.loc[ei,'Close'] 
                    if apa.loc[ei,'Share'] == 1 
                    else 0 for ei in apa.index]
```

```python
apa['Profit'].plot()
plt.axhline(y=0,color='red')
```

```python
apa['Wealth'] = apa['Profit'].cumsum()
```
