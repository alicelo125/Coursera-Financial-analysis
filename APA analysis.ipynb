```python
apa['Price1'] = apa['Close'].shift(-1)
# close price of tomorrow
```

```python
apa['PriceDiff'] = apa['Price1']-apa['Close']
apa.head()
```

```python
#Daily return = price difference/ close price of today
apa['Return'] = apa['PriceDiff']/apa['Close']
apa.head()
```

```python
# Direction (PriceDiff>0 up, <0 down)
apa['Direction'] = [1 if apa.loc[ei,'PriceDiff']>0 else -1                          for ei in apa.index]
apa.head()
```

```python
# Moving average 
apa['MA3'] = (apa['Close']+apa['Close'].shift(1)+apa['Close'].shift(2))/3
```

# Strategy 
### When MA10 > MA50 buy and hold one share of stock

```python
apa['MA10'] = apa['Close'].rolling(10).mean() # fast signal
apa['MA50'] = apa['Close'].rolling(10).mean() # slow signal
# rolling window to
```

```python
apa['Close'].plot(legend=True,figsize=(20,8))
apa['MA50'].plot(legend=True,figsize=(20,8))
apa['MA10'].plot(legend=True,figsize=(20,8))
```

```python
apa['Share'] = [1 if apa.loc[ei,'MA10']>apa.loc[ei,'MA50'] else 0 
                     for ei in apa.index]
apa['Profit'] = 
```

```python
apa['Close1'] = apa['Close'].shift(-1)
apa['Profit'] = [apa.loc[ei,'Close1']-apa.loc[ei,'Close'] 
                    if apa.loc[ei,'Share'] == 1 
                    else 0 for ei in apa.index]
```

```python
apa['Profit'].plot()
plt.axhline(y=0,color='red')
```

```python
apa['LogReturn'] = np.log(apa['Close']).shift(-1) - np.log(apa['Close'])
apa['LogReturn'].hist(bins=50)
# Log daily return of APA
```

```python
from scipy.stats import norm
```

```python
# Probability Density Function & Cumulative Distribution Function
density = pd.DataFrame()
density['x'] = np.arange(-4,4,0.001)
density['pdf']=norm.pdf(density['x'],0,1)
density['cdf']=norm.cdf(density['x'],0,1)
```

```python
plt.plot(density['x'],density['pdf'])
```

```python
plt.plot(density['x'],density['cdf'])
```

```python
mu = apa['LogReturn'].mean()
sigama = apa['LogReturn'].std(ddof=1)
```

```python
denApa= pd.DataFrame()
denApa['x']=np.arange(-0.1,0.1,0.001)
denApa['pdf']=norm.pdf(denApa['x'],mu,sigama)
```

```python
plt.ylim(0,20)
plt.plot(denApa['x'],denApa['pdf'])
plt.fill_between(x=np.arange(-0.1,0.1,0.0001),
                 y2=0,
                 y1=norm.pdf(np.arange(-0.1,0.05,0.0001,mu,sigama),  
                 facecolor='pink',
                 alpha=0.5))
```

```python
prob_return1 = norm.cdf(-0.05,mu,sigama)
print('the probability is',prob_return1)
```
